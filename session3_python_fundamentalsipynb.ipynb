```python
##accessing the element/ character of the string
##indexing

name ='tanzila'
#positive indexing
print(name[0])
#negative indexing
print(name[-1])
print(name[-2])
```

```python
## slicing
##accessing sub-string of a string

name ='tanzila'
print(name[0:])

#last digit in python is excluded
print(name[1:4])

# 0 to add digit
print(name[:3])

#print the whole string by using : sign
print(name[:])

#reverse the string
print(name[::-1])

#step size in slicing
print(name[1:6:2])
```

```python
##Slicing
#negative slicing
s ='hello world'
print(s[5:0:-2])

print(s[-5:])
print(s[-1:-6:-1])
```

```python
#removing string
k='tanzila'
del(k)
#k did not found as it deleted
print(k)
```
