##lets creates a function(with docstring)

```python

def is_even(num):
  """
  this is simple function to check the numbers is even or odd.
  input: any valid integer
  output: odd or even
  """
  if type(num) == int:
    if num%2==0:
      return "even"

    else:
      return "odd"
  else:
    print('enter valid data type')
```

##*args and kwargs

*args allow to pass the non keyword arguments of veriable length

*kwargs allow to pass the  keyword arguments of veriable length.

keywords arguments means that they contain key value pair like a python dictionary


```python
#*args allow to pass the non keyword arguments of veriable length


def multiply(*args):
  product =1
  for i in args:
    product=product*i
  return product
```

```python
multiply(1,2,34,5)
```

```python
#**kwargs allow to pass the  keyword arguments of veriable
def display(**kwargs):
  for key,value in kwargs.items():
    print(key, "-->",value)
```

```python
display(pakistan="islamabad",india='delhi',bangaladash='dhaka')
```

##Higher order function
Function that take another function as an input or give a function in its output is called higher order function

###map
###filter
###reduse

these are higher order function that take another or lambda function as an input


###map

```python
def square(x):
  return x**2

def transform(f,L):
  product=[]
  for i in L:
    product.append(f(i))
  print(product)


```

```python
transform(square,[1,2,3,4])
```

```python
list(map(lambda x:x**3, [1,2,3,4]))
```

```python
list(map(lambda x: 'even' if x%2==0 else 'odd',[1,2,3,4,5]))
```

###filter

filter items on the base of some condition

```python
#number greater than 5
list(filter(lambda x:x>5 ,[1,2,5,6,7,8,9,10]))
```

```python
#fetch fruits name starting from a
fruits =['apple','guava','pineapple',]
list(filter(lambda x:x.startswith('a'),fruits))
```
