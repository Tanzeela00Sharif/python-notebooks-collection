# Tuple

```python
#creating tuple ---> a=()
#empty
t1=()
print(t1)

#single item tuple
t2=(2,)
print(t2)
print(type(t2))

#homogeneous tuple
t3=(1,2,3,4)
print(t3)


#heterogeneous tuple
t4=(1,2.3,'apple',False)
print((t4))

#2D tuple
t5=((1,2),('apple','banana'))
print(t5)

#tuple creation by using type conversion

t6=('hello')
print(tuple(t6))
```

##Access the items of tuple

indexing

slicing

```python
#positive indexing
t4=(1,2.3,'apple',False)
print(t4[2])

#Negative indexing
print(t4[-3])

##positive slicing
print(t4[0:3:2])


#Negative slicing
print(t4[-3:-1])

#reverse the tuple
print(t4[::-1])


#2D tuple
t5=((1,2),('apple','banana'))
print(t5[1][0])
```

##tuple Editing

##Frozen set

All function of set work on frozenset. like union , intersections, difference and symetric difference.

```python
f1=frozenset([1,2,34])
print(type(f1))
print(f1)
```

```python
##frozenset are immutable vesion of set.all read fuction work but edit/wite function in it does not work.
f1=frozenset([1,2,34])
f2=frozenset([1,2,3,4,7,8,9])
print(f1|f2)
print(f1&f2)
print(f1-f2)
print(f1^f2)
```

frozen sets are used when the application only allows read only function. because it is immutable set version. set is use where read and write both  are use by the user.

```python
#2D frozen set
fs=frozenset([1,2,3,3,frozenset([7,8,9])])
fs
```

##dictionary comprehension

###{key:value  for var in iterable}

```python
{i:i**2 for i in range(1,11)}
```

```python
distance ={'gujrat':1000, 'lahore':2000, 'karachi':3000}
{key:value*0.72 for (key,value) in distance.items()}
```

```python
#zip
days = ['sunday','Monday','Tuesday']
tem=[12,13,14]
{i:j for (i,j) in zip(days,tem) }
```

```python
days = ['sunday','Monday','Tuesday']
tem=[12,13,14]
print(list(zip(days,tem)))
```

```python
products = {'phone':100,'laptop':200, 'charger':0}
{key:value for (key,value) in products.items() if value>100}
```
