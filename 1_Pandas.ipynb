##pandas series

pandas series is the 1D column in a table.it is a 1D array holding data of any type.

##Importing pandas

```python
import numpy as np
import pandas as pd
```

##Series from list

```python
#from string list
country=['pakistan','india','iran','nepal']
pd.Series(country)
```

```python
#from integers list
marks =[12,34,54,67,89]
pd.Series(marks)
```

```python
##custom index
marks =[12,34,54,67,89]
subject=['english','urdu','math','math','science']
pd.Series(marks,index=subject)
```

```python
##custom name of series data
marks =[12,34,54,67,89]
subject=['english','urdu','math','math','science']
pd.Series(marks,dtype=np.int32,index=subject,name='tanzila k marks')
```

##Series from dict

```python
#keys become index while value become the series data
marks={
    'urdu':13,
    'english':56,
    'math':51
}
marks_series=pd.Series(marks, name='marks of tanzila')
marks_series
```
