##Stacking
###shape must be same
use to join the different arrays with same shape horizontly and vertically

```python
a5=np.arange(12).reshape(3,4)
a6=np.arange(12,24).reshape(3,4)
a6
```

```python
#horizontal stacking
a5=np.hstack((a5,a6))
a5
```

```python
#Vertical stacking
np.vstack((a5,a6,a5))
```
