# Operators in python

 Arithmetic Operators

Used for basic mathematical operations.

 + Addition → 2 + 3 = 5

- Subtraction → 5 - 2 = 3

* Multiplication → 3 * 2 = 6





/ Division → 6 / 2 = 3.0

// Integer Division → 7 // 2 = 3

% Modulus → 7 % 2 = 1

** Exponent → 2 ** 3 = 8

```python
print(7+10)
print(7/10)
print(7-10)
print(7//10)
print(10%7)
print(7**2)
```

2. Comparison (Relational) Operators


Used to compare values (returns True/False).


== Equal to

!= Not equal to

> Greater than

< Less than
>= Greater than or equal to

<= Less than or equal to

```python
print(7==10)
print(7>10)
print(7<10)
print(7!=10)
print(7<=10)
print(7>=10)
```

3. Logical Operators

Used to combine conditions.

and → True if both are true

or → True if at least one is true

not → Reverses the result

```python
print(2&3)
print(2|3)
print(2^3)
```

5. Assignment Operators

Used to assign values.

= Assign

+= Add and assign

-= Subtract and assign

*= Multiply and assign

/= Divide and assign

%= Modulus and assign

**= Power and assign

```python
a=7
a+=7
print(a)

b=12
b-=7
print(b)

c= 13
c*=7
print(c)

d=14
d/=7
print(d)

f=10
f//=7
print(f)

j=7
j%=7
print(j)
```

7. Membership Operators

Check if a value exists in a sequence.

in

not in

```python
print('t' in 'tanzila')
print(2 not in [1,2,3,4])
print (7 in (1,2,3,4,7))
```

```python

##guessing game
import random
jackpot= random.randint(1,100)

guss=int(input('enter the guss'))
Counter=1
while guss!=jackpot:
  if  guss<jackpot:
    print('guss higher')
  else:
    print('guss lower')
  guss=int(input('enter the guss'))
  Counter+=1
else:
  print('correct guss')
  print('attempts',Counter)
```

# Nested loop

```python
#print unique pairs
for i in range (1,5):
  for j in range(1,5):
    print(i,j)
```

```python
rows = int(input('enter the row number:'))
for i in range(1, rows+1):
  for j in range(1,i+1):
    print('*',end='')
  print()
```

```python
rows = int(input('enter the row number:'))
for i in range(1, rows+1):
  for j in range (1,i+1):
    print(j, end='')
  for k in range(i-1,0,-1):
    print(k, end='')
  print ()
```

# Loop control statements

break ----> loop breaks when spacific condition meet, use for searching

continue  ----->  loop skip specific thing and than continue

pass ----->  use in code when we dont have code and write code in future

```python
for i in range(1,11):
  if i== 5:
    break
  print(i)
```

```python
#break
lower= int(input('enter lower number'))
upper= int(input('enter upper number'))

for i in range(lower,upper+1):
  for j in range(2,i):
    if i%j==0:
      break

  else:
    print(i)
```

```python
#continue for skiping speciic things
#use to check the out of stock items in e-commerce store
for i in range(1,11):
  if i== 5:
    continue
  print(i)
```

```python
#use for error handling when you not know the answer
for i in range(1,10):
  pass
```
